# Actualización de inventario
Problema tomado de [freeCodeCamp](https://www.freecodecamp.org/)

**Descripción**

Compare y actualice el inventario almacenado en una matriz 2D contra una segunda matriz 2D de una entrega nuevva. Actualice las cantidades de artículos actuales en el inventario extistente (en arr1). Si un artículo no puede ser encontrado, agregar el item nuevo y la cantidad en la matriz del inventario. La matriz de inventario retornada debe estar ordenada alfabéticamente por artículo.

# Paso 1: Definir las matrices

Se crean las matrices **curInv** y **newInv** para almacenar el inventario actual y la actualizacion.

```python
curInv = [
  [21, "Bowling Ball"],
  [2, "Dirty Sock"],
  [1, "Hair Pin"],
  [5, "Microphone"]
]

newInv = [
  [2, "Hair Pin"],
  [3, "Half-Eaten Apple"],
  [67, "Bowling Ball"],
  [7, "Toothpaste"]
]
```

# Paso 2: Crear función para ordenar listas

Se crea la función **sortArray** la cual recibe como parámetro cada artículo de la lista (**item**) y retorna la columna que se va a ordenar.

```python
def sortArray(item: list) -> list:  # Línea 1
  return item[1]                    # Línea 2
```

## Explicación función **sortArray**

Se utiliza para ordenar matrices con 2 o mas columnas, y se debe pasar como parametro **key** de la función sort(), ejemplo: **matriz.sort(key = sortArray)**.

La posición que se retorna (*línea 2*) corresponde a la columna respecto a la cual se quiere ordenar la matriz, 0->Columna 1, 1->Columna 2, n->Columna n-1.

# Paso 3: Crear función para actualizar inventario

Se crea la función **updateInventory** la cual recibe como parámetros **arr1 -> curInv** y **arr2 -> newInv**, y retorna el **arr1**

```python
def updateInventory(arr1: list, arr2: list) -> list:  # Línea 1
  index = None                                        # Línea 2
  for i in range(len(arr2)):                          # Línea 3
    for j in range(len(arr1)):                        # Línea 4
      if arr1[j][1] == arr2[i][1]:                    # Línea 5
        index = (i, j)                                # Línea 6
        break                                         # Línea 7
    if index == None:                                 # Línea 8
      arr1.append(arr2[i])                            # Línea 9
      index = None                                    # Línea 10
    else:                                             # Línea 11
      arr1[index[1]][0] += arr2[index[0]][0]          # Línea 12
      index = None                                    # Línea 13
  arr1.sort(key=sortArray)                            # Línea 14
  return arr1                                         # Línea 15
```

## Explicación función **updateInventory**

Se comienza inicializando la variable index (*línea 2*), la cual se encarga de almacenar los indices de los artículos en común de las 2 matrices, luego se itera por cada elemento del **arr2** (*línea 3*), y se anida la iteración por cada elemento del **arr1** (*línea 4*), al estar anidados ambos bucles **for**, cada elemento del **arr2** es comparado con todos los elementos del **arr1** para verificar si existe en el **arr1** (*linea 5*). Si existe, se almacenan ambos indices en la variable **index** (*línea 6*)  y se rompe el ciclo (*línea 7*), al final del ciclo for anidado, se verifica si la variable **index** mantiene su valor de *None* (*línea 8*), lo cual significa que el elemento del **arr2** no está en el **arr1**, se agrega el elemento al **arr1** (*línea 9*) y se vuelve a asignar el valor de *None* a la variable **index** (*línea 10*), si no mantiene el valor de *None*, es que se encontró una coincidencia, y se suma la cantidad del elemento del **arr2** en la cantidad del elemento del **arr1** (*línea 12*), y se vuelve a asignar el valor de *None* a la variable **index** (*línea 13*), finalmente se ordenan los artículos del **arr1** en orden alfabético (*linea 14*) y se retorna el **arr1** para finalizar la función (*línea 15*).
